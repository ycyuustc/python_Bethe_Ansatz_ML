<p>
	<span style="color:#003399;font-size:16px;">The basic settings:</span>&nbsp;
</p>
<p>
	Num: the discrete point number
</p>
<p>
	L: period in the X space
</p>
<p>
	$\Delta x = \frac{L}{N}$
</p>
<p>
	$\Delta k = \frac{2\pi}{L}$
</p>

```python
Num = 200
L = 20.0
delta_x = L/Num
delta_k = 2*pi/L

vx = np.arange(Num) * delta_x
vk0 = np.arange(Num) * delta_k
Numkeep = int(Num/2) + 1
Nummove = Num - Numkeep
ind_zero = Nummove 

vkkeep = vk0 [0:Numkeep]
vkmove = vk0 [Numkeep:Num] - 2*pi*Num/L
vk = np.concatenate((vkmove, vkkeep), axis=0)
print(vk[ind_zero])

plt.plot(np.arange(Num),vk, '*')
```

```python
v_ind = np.arange(Num).astype(int)
tv = np.ones(Num*Num).astype(int)
tm = np.dot(v_ind.reshape([Num,1]), tv.reshape([1,Num*Num]))
m3indx = tm.reshape([Num, Num, Num])
m3indy = m3indx.transpose([1, 0, 2])
m3indz = m3indx.transpose([2, 1, 0])

tm = np.dot(vk.reshape([Num,1]), tv.reshape([1, Num*Num]))
m3kx = tm.reshape([Num, Num, Num])
m3ky = m3kx.transpose([1, 0, 2])
m3kz = m3kx.transpose([2, 1, 0])

m3energy = np.zeros([Num, Num, Num])
```

<p>
	<span style="color:#EE33EE;font-size:24px;"><span style="color:#337FE5;">Notice that we sort all the vector and tensor only one time by </span><span style="color:#EE33EE;"> <em><u>vsort</u></em></span><span style="color:#003399;"><span style="color:#337FE5;">, which is the sort of the energy spectrum.&nbsp;</span><span style="color:#337FE5;"></span></span></span>
</p>

```python
m3energy = m3kx**2 + m3ky**2 + m3kz**2
num_total = np.prod(m3energy.shape)
venergy = m3energy.reshape(num_total)

vindx = m3indx.reshape(num_total)
vindy = m3indy.reshape(num_total)
vindz = m3indz.reshape(num_total)
vkx = m3kx.reshape(num_total)
vky = m3ky.reshape(num_total)
vkz = m3kz.reshape(num_total)

vsort = venergy.argsort()
v_sort_energy = venergy[vsort]
v_sort_indx = vindx[vsort]
v_sort_indy = vindy[vsort]
v_sort_indz = vindz[vsort]
v_sort_kx = vkx[vsort]
v_sort_ky = vky[vsort]
v_sort_kz = vkz[vsort]

energy_cut = max(vk**2)
ind_cut = np.min(np.where(v_sort_energy>energy_cut))
m2_sort_k = np.zeros([ind_cut, 3])
m2_sort_k[:, 0] = v_sort_kx[np.arange(ind_cut)]
m2_sort_k[:, 1] = v_sort_ky[np.arange(ind_cut)]
m2_sort_k[:, 2] = v_sort_kz[np.arange(ind_cut)]
m2_sort_ind = np.zeros([ind_cut,3])
m2_sort_ind[:, 0] = v_sort_indx[np.arange(ind_cut)]
m2_sort_ind[:, 1] = v_sort_indy[np.arange(ind_cut)]
m2_sort_ind[:, 2] = v_sort_indz[np.arange(ind_cut)]
```

```python
plt.figure(figsize = (20,8), dpi = 80)
temp = 5
tv = v_sort_energy[np.arange(ind_cut)]
num_bins = int((np.max(tv) - np.min(tv))/temp)
plt.hist(tv, num_bins)
plt.grid(alpha = 0.4)
plt.show()
```
