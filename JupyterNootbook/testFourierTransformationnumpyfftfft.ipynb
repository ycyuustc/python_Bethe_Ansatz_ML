```python
import matplotlib.pyplot as plt
import numpy as np
pi = np.pi

from IPython.display import Image
PicPath='C:/Users/17613/Pictures/Jupyter/'
```

```python
# test the fast fourier transform

v_input = [1,3,4,7,9]
num = len(v_input)
v_x = np.arange(num)
v_k = np.arange(5)*(1.0/num)
[mx,mk] = np.meshgrid(v_x, v_k)
mt = np.exp(-1j*2*pi*mx*mk)
v_test = np.dot(mt,np.transpose(v_input))

v_ind = np.arange(5) + 1
v_result = np.fft.fft(v_input)
fig = plt.figure('f(x)')

ax1 = fig.add_subplot(211)
ax1.plot(v_x, np.real(v_input), '-g*')
ax2 = fig.add_subplot(212)
ax2.plot(v_x, np.imag(v_input), '-r*')

fig2 = plt.figure('f(k)')
ax1 = fig2.add_subplot(211)
ax1.plot(v_k, np.real(v_result), '-g*')
ax2 = fig2.add_subplot(212)
ax2.plot(v_k, np.imag(v_result), '-r*')

fig3 = plt.figure('ftest(k)')
ax1 = fig3.add_subplot(211)
ax1.plot(v_k, np.real(v_test), '-g*')
ax2 = fig3.add_subplot(212)
ax2.plot(v_k, np.imag(v_test), '-r*')

plt.show()
```

```python
v_input = [1,3,4,7,2]
num = len(v_input)
v_x = np.arange(num)
v_k = np.arange(5)*(1.0/num)
[mx,mk] = np.meshgrid(v_x, v_k)
mt = np.exp(1j*2*pi*mx*mk)
v_test = (1.0/num)*np.dot(mt,np.transpose(v_input))

v_ind = np.arange(5) + 1
v_result = np.fft.ifft(v_input)
fig = plt.figure('f(x)')

ax1 = fig.add_subplot(211)
ax1.plot(v_x, np.real(v_input), '-g*')
ax2 = fig.add_subplot(212)
ax2.plot(v_x, np.imag(v_input), '-r*')

fig2 = plt.figure('f(k)')
ax1 = fig2.add_subplot(211)
ax1.plot(v_k, np.real(v_result), '-g*')
ax2 = fig2.add_subplot(212)
ax2.plot(v_k, np.imag(v_result), '-r*')

fig3 = plt.figure('ftest(k)')
ax1 = fig3.add_subplot(211)
ax1.plot(v_k, np.real(v_test), '-g*')
ax2 = fig3.add_subplot(212)
ax2.plot(v_k, np.imag(v_test), '-r*')

plt.show()
```
